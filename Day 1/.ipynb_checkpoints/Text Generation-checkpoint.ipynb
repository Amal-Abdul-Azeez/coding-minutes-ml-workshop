```python
text = "the man was ....they...then.... then ... the ... the  "

# X is the sequence of 'K = 3' and Y is predicted character or K+1 the character

X      Y     Freq
the    " "    3
he     "m"    1
e m.   "a"    1
the    "n"    2
the    "y"    1
the    "i"    1
man    "_"    1
```

```python
def generateTable(data, k=4):
    
    # empty dictionary
    T = {}
    
    for i in range(len(data)-k):
        X = data[i:i+k]
        Y = data[i+k]
        #print("X  %s and Y %s  "%(X,Y))
        
        if T.get(X) is None:
            T[X] = {}
            T[X][Y] = 1
        else:
            if T[X].get(Y) is None:
                T[X][Y] = 1
            else:
                T[X][Y] += 1
    
    return T
```

```python
T = generateTable("hello hello helli helly")
```

```python
T
```

```python
def convertFreqIntoProb(T):     
    for kx in T.keys():
        s = float(sum(T[kx].values()))
        for k in T[kx].keys():
            T[kx][k] = T[kx][k]/s
                
    return T
```

```python
convertFreqIntoProb(T)
```
