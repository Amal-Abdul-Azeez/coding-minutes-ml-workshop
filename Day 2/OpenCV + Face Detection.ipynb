```python
import cv2 # opencv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = plt.imread('group_photo.jpeg')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img_copy = img.copy()
```

```python
model = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces = model.detectMultiScale(img)
```

```python
faces.shape
```

```python
faces
```

```python
x,y,w,h = faces[0]
```

```python
img_copy = cv2.rectangle(img_copy, (x, y) , (x+w, y+h), (255,0,0), 2 )
```

```python
plt.imshow(img_copy)
```

```python
for one_face in faces:
    x,y,w,h = one_face
    img_copy = cv2.rectangle(img_copy, (x, y) , (x+w, y+h), (255,0,0), 2 )
```

```python
plt.imshow(img_copy)
```

```python
plt.imsave("detected.jpg", img_copy)
```
