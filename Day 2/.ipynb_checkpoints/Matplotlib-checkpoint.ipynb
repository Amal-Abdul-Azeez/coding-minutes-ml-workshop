# Scatter plot

```python
X = np.round(np.random.randn(100, 1), decimals=1)
```

```python
noise = np.random.randn(100,1)*10 + 10
```

```python
np.random.randn(100).shape
```

```python
Y = X*10 + 20 + noise
```

```python
X.shape
```

```python
Y.shape
```

```python
X[:5]
```

```python
Y[:5]
```

```python
plt.scatter(X, Y)
plt.scatter(-1, 50, s = 300, marker='*')
plt.xlabel("Revenue")
plt.ylabel("Profit")
```

## Pie chart

```python
marks = np.array([91, 55, 78, 82])
subjects = ["Physics", "Maths", "Chem", "CS"]
```

```python
plt.pie(x = marks, labels=subjects, )
plt.show()
```
